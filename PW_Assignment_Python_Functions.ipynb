6. Explain the concept of generators in Python and how they are defined ?

Ans:

Generators are used for efficient memory usage. They are generally used for storing or accessing data one by one rather than storing it all in a memory and then accessing it.

They can be defined as a function or as an expression :    

```
As a function :
def gen(n):
  for i in n:
    yield i
```

```
As an Expression :

gen = (i for i in range(1,100000))

next(gen)
next(gen)
```

7. What are the advantages of using generators over regular functions?

Ans:

- Generator functions returns one value at a time whereas regular function executes all at once and returns it

- Generator functions can be paused and resume whenever needed until the function gets exhausted whereas it is not possile with regular functions

- Generator functions are used for larger data sets as it returns one output value at time and is very efficient

- Generator functions are memory efficient cause they store only one value in the memory at a time where as regular functions stores all the data in the memory which is returned

10. What is the difference between `map()`, `reduce()`, and `filter()  functions in Python?

Ans:

All the three are bulit in functions in python.

- Map --> applies the function on each element of the iterable and returns a iterator like a generator function

- reduce --> applies the function on each element and returns a single value. the function used in the reduce must be passed with 2 arguments

- filter --> applies the function on each element of iterable and reutnrs the values that matches with the criteria defined in the function

Example:

map >

tup = (1,2,3,4,5)

print(list(b))

reduce >

from functools import reduce

tup = (1,2,3,4,5)

print(reduce(lambda x, y: x if x > y else y, tup))


filter >

from functools import reduce

tup = (1,2,3,4,5)

print(list(filter(lambda x: x == 2, tup)))

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list ?
from functools import reduce

def sumOfeven(*args):
  return reduce(lambda x,y: x + y, list(filter(lambda x: x % 2 == 0, list(args))))

print(sumOfeven(2,4,6,8))
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class iterator:

  def __init__(self, terms):
    self.terms = terms

  def fib(self):
      if self.terms == 0:
         return 0
      elif self.terms == 1:
          return 1
      else:
          a = 0
          b = 1
          for i in range(0, self.terms):
              if i == 0:
                current = a + a
              elif i == 1:
                current = a + b
              else:
                current = a + b
                a = b
                b = current
              yield current

gen = iterator(6)

fiba = gen.fib()

print(next(fiba))
print(next(fiba))
print(next(fiba))
print(next(fiba))
print(next(fiba))
print(next(fiba))
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent ?


def gene(num):
  for i in range(0, num):
    yield 2**i

pow = gene(10)

print(next(pow))
print(next(pow))
print(next(pow))
print(next(pow))
print(next(pow))
print(next(pow))
````
