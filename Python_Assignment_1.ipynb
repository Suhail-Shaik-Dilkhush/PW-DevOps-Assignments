3.Compare and contrast mutable and immutable objects in Python with examples?

**Mutable Objects** :

1. the objects are data types in which data can be modified.

2. data in these objects can be replaced or deleted or modified with any value

3. these are used in the use cases where data modification is required.

Ex : storing price of an item, storing name of student in records

Some of them are ->

=> lists - data in a list can be declared perviously and can be alterted whenever wanted

=> dictionaries - can add or remove or modify any key value pair if required

**Immutable Objects** :

1. The objects of data type in which data cannot be altered once declared

2. after defining a data in this type of object, it cannot be modified, removed or replaced

3. used in use case where the data modification is not required

Ex : student id number, employee id number

some of them are ->

=> string - strings cannot be altered. even we try to use the methods like slice to modify, it generally returns a new string and not the original string

=> Tuples - tuples are like lists, but the data in the lists once declared cannot be altered. But if the Tuple has a list data type in one of its elements, then that list elements can be altered

- conditional statemets or blocks can be used to perform an operation or task based upon the decision of "true" or "false"

- if else, match case

- can use break and continue in these conditions to skip an iteration or break out of a loop.

```
# Example of If Else Block

a = int(input())

if (a % 2 == 0){

  print(f"{a} is an even number")
}
else {

  print(f"{a} is an Odd number")
}

#Example of Match Case

def even_odd(numb):
  match numb % 2:
    case 0:
      return f"{numb} is an even number"
    case _:
      return f"{numb} is an Odd number"
     
print(even_odd(5))

```
