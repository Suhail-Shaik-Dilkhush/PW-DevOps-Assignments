17. Describe the time complexity of accessing elements in a dictionary ?

Ans: Dictionary uses key-value pairs, in which keys are strings or other immutable data types. Dictionary converts these keys into hash values and uses them for performing searching or accessing the data. The hash value look ups are very fast and gives the time complexity of O(1).
