Task 1: File Management Script

Write a Bash script thatg
- Creates a directory named "backup" in the user's home directory^
- Copies all .txt files from the current directory into the "backup" directory^
- Appends the current date and time to the filenames of the copied files.

Solution :
```
#!/bin/bash

mkdir $HOME/Backup #creating a the backup directory in users home direc

for file in ./*.txt;
do
  name=$(basename $file)
  current_date=$(date +%Y-%m-%d-%H-%M)
  cp $file $HOME/Backup/$name-$current_date.txt
done

```

Task 2: System Health Check

Create a script thatg
- Checks the systemâ€™s CPU and memory usage^
- Reports if the CPU usage is above 80% or if the available memory is below 20%^
- Logs the results to a file named system_health.log

Solution :

```
#!/bin/bash

cpu=$(top -b -n 1 | awk "/%Cpu/" | awk "{ print $ 8 }") #fetching idle time of cpu using the top command

mem=$(free | awk '/Mem/' | awk '{print($7/$2 * 100)}') #calculating free memory using formula - (available memory/ total memory)*100


if (( $(echo "100 - $cpu > 80" | bc -l) )); then
   echo "Cpu utiliztation is more than 80% - find cpu idle time : $cpu" >> system_health.log
fi

if (( $( echo "$mem > 20" | bc -l) )); then
   echo "Avaialable free memory is below 20% - current limit - $mem - at $(date)" >> system_health.log
fi


```

Task 3: User Account Management

Write a script thatg
- Reads a list of usernames from a file (e.g., user_list.txt)^
- Creates a new user for each username^
- Generates a random password for each user and saves the username and password to a file named
credentials.txt.

Solution :

```
user_list="/home/suhail/user_list.txt"

echo $user_list

for user in $(cat $user_list);
do
password=$(openssl rand -base64 12) #using openssl to generate a random password of 12 chars

useradd -m -d /home/$user -s /bin/bash -p$password $user

echo "$user:$password" > credentials.txt

done

```

Task 6: Automated Software Installation

Write a script thatg
- Reads a list of software package names from a file (e.g., packages.txt)^
- Installs each package using the appropriate package manager (apt, yum, etc.)^
- Logs the installation status of each package

```

packages=$HOME/packages.txt

distro=$(sed -n 's/ID_LIKE=//p' /etc/os-release) #fetching the distro name from the ID_LIKE attribute in os-realase file

while read package;
do

#first letter of ditro skipped cause it can be capital or small

if [[ "$distro" == *"buntu"* || "$distro" == *"ebian"* ]];then

   sudo apt install $package -y

elif [[ "$distro" == *"entOS"* || "$distro" == *"HEL"* || "$distro" == *"ocky"* || "$distro" == *"lmaLinux"* ]]; then

   sudo dnf install $package -y || sudo yum install $package -y

elif [[ "$distro" == *"rch"* || "$distro" == *"anjaro"* ]]; then

  sudo pacman -Syu $package

elif [[ "$distro" == *"pen"* || "$distro" == *"SLE"* ]]; then

  sudo zypper install -y $package

elif [[ "$distro" == *"lpine"* ]]; then

  sudo apk add $package -y
else
  echo "no supported distro found : $distro"

fi

done < $packages


```
