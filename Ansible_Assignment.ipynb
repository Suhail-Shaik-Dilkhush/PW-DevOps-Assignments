1. Write Ansible playbooks to automate the setup and configuration of a web
server (e.g., Apache or Nginx)?

```python
"""
- name: Nginx setup
  hosts: webserver
  become: yes
  vars:
    packages:
      ubuntu: nginx
      debian: httpd

  tasks:
    - name: installing the package
      package:
        name: "{{ packages[ansible_distribution] }}"
      state: present

    - name: copying the nginx configuration files to the host
      template:
        src: ./nginx.conf.js
        dest: /etc/nginx/conf.d/nginx.conf
      when: ansible_distribution == {{ packages[ubuntu] }}

    - name: copying the httpd configuration file to the host
      template:
        src: ./httpd.conf.js
        dest: /etc/apache2/conf.d/httpd.conf
      when: ansible_distribution == {{ packages[debian] }}

"""
```
